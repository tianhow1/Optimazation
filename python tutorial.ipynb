Example: Plotting the feasible region of an optimization problem

$\min_{x \in \mathbb{R}^2} (x_1 - 2)^2 + (x_2 - 1)^2$ s.t $x_2 - x_1^2 \ge 0$, $x_1 + x_2 \le 2$

```python
import matplotlib.pyplot as plt
import numpy as np
```

Imports matplotlib (plotting library) and numpy (package for working with mathematical functions)

Plot the feasible region

We're plotting 
$x_2 \ge x_1^2$ and $x_2 \le 2 - x_1$

```python
%matplotlib inline
x = np.linspace(-3, 3, 2000) # generates a 2000 xs 1 vector equally spaced from -3 to 3
y1 = x*x           # x_2 = x_1^2: multiplying two vectors together is done elementwise. 
y2 = 2 - x         # x_2 = 2 - x_1: subtract or add a number to a vector, that's also done elementwise.
fig, ax = plt.subplots()
ax.plot(x, y1, 'k--', label=r'$x_2 - x_1^2\geq 0$, above the curve')
ax.plot(x, y2, 'k:', label=r'$2 - x_1 -x_2\geq 0$, below the line')
plt.xlim((-3, 3))
plt.ylim((0, 6))
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.fill_between(x, y2, y1, where=y2>y1, color='grey', alpha=0.5)
plt.title('Feasible Region')
legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.show()
```

## Mathematical operations in python

```python
print(3+2) # Addition
print(3-2) # Subtraction
print(3*2) # Multiplication
print(3/2) # Division # output 1.5

print(3//2) # Integer division: 3//2 = 1 (get rid of decimal part)
print(3**2) # Exponentiation: 3 **2 = 9, 3^2
print(3 % 2) # Modulus: 3 % 2 = 1,
print(3^2) # Bitwise XOR: don't confuse this with 3**2: 3^2 = 1
```

If you want to evaluate the objective function (in this case it's $(x_1 - 2)^2 + (x_2 - 1)^2$)

Function syntax:

def functionname(var1, var2):

    write any statements here

    return the value you want to output

```python
def func(x1, x2):
    x1 = 2
    return (x1 - 2)**2 + (x2 - 1)**2
# anything in the function needs to be indented
```

Anonymous functions: If you have a simple function, you can write it in one line using anonymous functions:

functionname = lambda var1, var2: whatever you want to return

```python
anonymous = lambda x1, x2: (x1 - 2)**2 + (x2 - 1)**2
```

Call the function with functionname(param1, param2)

```python
print(func(2, 2)) # (2 - 2)^2 + (2 - 1)^2 = 1
print(anonymous(3,2)) # (3 -2)^2 + (2 - 1)^2 = 2
```

### Example: Plotting a function of two variables

Consider $f(x,y) = x^2 + 2y^2 + 4x + y + 6$. Below, we graph this function in 3D using the `mpl_toolkits.mplot3d` module provided by `matplotlib`.

```python
import numpy as np  # arrays
import matplotlib.pyplot as plt  # 2d plotting
from mpl_toolkits.mplot3d import axes3d  # 3d plotting

f = lambda x, y: x**2 + 2*y**2 + 4*x + y + 6
f(1,-1)
```

```python
X = np.linspace(-10,10,300)  # 300 evenly spaced points on x-axis [-10,10]
Y = np.linspace(-10,10,300)  # 300 evenly spaced points on y-axis [-10,10]
Xmesh, Ymesh = np.meshgrid(X,Y)
print(X)
print(Y)
```

```python
print(Xmesh)
print(Ymesh)
```

```python
# Mathematical operations in definition of f act componentwise on NumPy arrays
X**2
# 2*Y
# X+Y
```

```python
f(X,Y)  # evaluate at (X[0],Y[0]), then (X[1],Y[1])), and so on
```

```python
Xmesh, Ymesh = np.meshgrid(X,Y)  # creates 300x300 grid of points defined by X and Y above
print(Xmesh.shape)
print(Xmesh[:5,:5])
print(Ymesh[:5,:5]) # grid [-10,10] x [-10,10]
```

```python
Xmesh
```

```python
Z = f(Xmesh, Ymesh)  # evaluate f at each of the 300x300 points
print(Z[:5,:5])
Z.shape
```

```python
fig = plt.figure(figsize=(10, 6))
# feel free to play with the arguments below to get a feel for 3D plotting:
#     elev is angle above xy-plane, azim is angle around z-axis
#     Ex: elev=0, azim=0 shows yz-plane with positive x-axis out of screen
ax = plt.axes(projection='3d', elev=15, azim=-60)
ax.plot_surface(Xmesh, Ymesh, Z, cmap='jet', alpha=0.5, vmax=250)
ax.set(xlabel='x', ylabel='y', zlabel='z', xlim=[-10,10], ylim=[-10,10], zlim=[0,256])
plt.show()
```
